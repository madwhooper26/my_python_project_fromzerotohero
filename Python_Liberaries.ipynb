```python
"""Let say your the teacher and you have different student records containing id of a student and the mark
list in each subjects where different students have taken different number of subjects. All thses records are 
in hard copy. you want to enter all the data in the computer and want to compute the average marks of each student 
and display"""

def getDataFromUser():
    D = {}
    while True:
        studentid = input("enter the studentId")
        marksList = input("enter the marks seperated by comma: ")
        moreStudents = input('enter "no" for quit isertion: ')
        if studentid in D:
            print("the studentId already inserted")
        else:
            D[studentid] = marksList.split(",")
        if moreStudents.lower() == "no":
            return D
    
            
            
    
    
```

```python
studentData = getDataFromUser()
```

```python
studentData
```

```python
def getAvgMarks(D):
    AvgMarksDictionary ={}
    for x in D:
        L = D[x]
        s= 0
        for marks in L:
            s+= int(marks)
        AvgMarksDictionary[x] = s/len(L)
    return AvgMarksDictionary
        
            
            
        
```

```python
avgM = getAvgMarks(studentData)
```

```python
for x in avgM:
    print("student :",x,"got avg Marks as:",avgM[x])
```
