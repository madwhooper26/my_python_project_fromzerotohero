""" Given a list of numbers i.e. [1,2,4,-5,7,9,3,2], make another list that
contains all the items in the sorted from min to max.
i.e. your result will the another list like in the sorted order
"""

```python
def findMin(L,startIndx):
    m = L[startIndx]
    idx = startIndx
    for i in range(startIndx, len(L)):
        x = L[i]
        if x<m:
            m = x
            idx = i
        else:
            pass
            
    return m,idx
```

```python
def swapValues(L,idx1,idx2):
    temp = L[idx1]
    L[idx1] = L[idx2]
    L[idx2] = temp
    return L
```

```python
L = [2,3,6,7]
L2 = swapValues(L,1,3)
print(L2)
```
